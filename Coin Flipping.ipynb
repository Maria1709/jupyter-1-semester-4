```python
# We will use numpy and scipy.stats to analyse flipping of a coin.
import numpy as np 
import scipy.stats as ss
```

```python
# We will use this for visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Flip a fair coin 1000 times! How many heads?
np.random.binomial (1000, 0.5, 1)
```

```python
# Number of heads, number of flips, and probability of a head.
ss.binom.pmf(521, 1000, 0.5) # This is the probability of getting 521 heads in 1000 flips.
```

```python
# Shows distribution of number of heads we get.
sns.distplot(np.random.binomial (1000, 0.5, 1000))
```

```python
# Flip an unfair coin 10 times and how many heads do you get? The probability of r heads when flipping an unfair coin
# n times is p(r,/n,p)=(nr)pr(1-p)(n-r)
np.random.binomial(10, 0.2, 1)
```

```python
# Result of the probability of heads fliping an unfair n times
(0.4)*(0.4)*(1.0-0.4)*(0.4)*(1.0-0.4)*(1.0-0.4)*(1.0-0.4)
```
